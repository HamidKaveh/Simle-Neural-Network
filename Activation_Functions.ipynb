# **Some Different Types of Activation Function Using in Machine Learning**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,50)
```

```python
text_font = {'family': 'consol', 'color':'#22242b', 'size':12}
title_font = {'family': 'consol', 'color':'#373f51', 'size':16, 'fontweight':'bold'}
```

## **Gaussian Error Linear Unit (GELU) function and its derivative:**

```python
# By using Math library:
import math
def Gelu(x):
    return (x/2)*(1+math.erf(x/(2**0.5)))

# Approximating the GELU:
def gelu(x):
    p = math.pi
    t = math.sqrt(2/p)
    i = t * (x + 0.044715*(x**3))
    return (0.5*x*(1 + ((np.exp(i)-np.exp(-i))/((np.exp(i)+np.exp(-i))))))


def Derivative_Gelu(x):
    delta = 0.5*math.tanh(0.0356774*(x**3)+0.797885*x) + (0.0535161*x**3 + 0.398942*x) * ((1/math.cosh(0.035677*x**2 + 0.797885*x))**2) + 0.5
    return delta
```

```python
y=[];
z=[];
for i in x:
    y.append(Gelu(i))
    z.append(Derivative_Gelu(i))

#figsize(10,5)
plt.grid()
plt.plot(x, y ,'#2a9d8f', linewidth=2.5)
plt.plot(x,z,'r', linewidth=2)
plt.xticks(np.arange(-4,3))
plt.yticks(np.arange(-0.5,4,0.5))
plt.title('Gelu Function', fontdict=title_font)
plt.annotate(r'$ \mathcal{Gelu(x)} = \mathcal{x}. \Phi(\mathcal{x}) = \frac{x}{2} (\mathcal{1} + \mathcal{erf} (\frac{x}{\sqrt{2}})) $', xy=(2.9, 3), xytext=(-4, 4), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative} = \Phi(\mathcal{x}) + \mathcal{x} \phi(\mathcal{x})$', xy=(0.5, 1), xytext=(-4.2, 1.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'));
```

## **Identity function & its derivative:**

```python
def Identity(x):
    return x

def Derivative_Identity(x):
    return 1
```

```python
y=[]
for i in x:
    y.append(Derivative_Identity(i))

#figsize(8,5)
plt.grid()
plt.plot(x, Identity(x) ,'#2a9d8f', linewidth=2.5)
plt.plot(x, y,'r', linewidth=2)
plt.xticks(np.arange(-4,+4,1))
plt.yticks(np.arange(-2,+4,1))
plt.title('Identity Function', fontdict=title_font)
plt.annotate(r'$Identity(x)=x$', xy=(-1.25, -1.25), xytext=(0.25, -3.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative} =1$', xy=(-1.25, 1), xytext=(-2, 3.25), fontsize=12, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Binary step function & its derivative:**

```python
def Binary_Step(x):
    if x<0:
        return 0
    else:
        return 1

def Derivative_Binary(x):
    if x != 0:
        return 0
    if x ==0:
        raise ValueError("Undefined")
```

```python
y=[];
z=[];
for i in x:
    y.append(Binary_Step(i))
    z.append(Derivative_Binary(i))

#figsize(8,5)
plt.grid()
plt.plot(x, y,'#2a9d8f', linewidth=2.5)
plt.plot(x, z,'r', linewidth=1)
plt.xticks(np.arange(-3,+3,1))
plt.yticks(np.arange(-1,+2,0.25))
plt.title('Binary Step Function', fontdict=title_font)
plt.annotate(r'$\mathcal{Binary Step}$', xy=(0, 0.5), xytext=(-3, 0.75), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative}$', xy=(1, 0), xytext=(1, -0.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Sigmoid(Logistic or Soft Step) function and its drivative:**

```python
def Sigmoid(x):
    return 1/(1+np.exp(-x))

def Derivative_Sigmoid(x):
    return (Sigmoid(x) * (1-Sigmoid(x)))
```

```python
#figsize(8,5)
plt.grid()
plt.plot(x, Sigmoid(x) ,'#2a9d8f', linewidth=2.5)
plt.plot(x, Derivative_Sigmoid(x),'r', linewidth=2)
plt.xticks(np.arange(-5,+5,1))
plt.yticks(np.arange(0,+1,0.25))
plt.title('Sigmoid Function', fontdict=title_font)
plt.annotate(r'$\sigma(x)=\frac{1}{1+e^{-x}}$', xy=(-0.10, 0.5), xytext=(-2, 0.75), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative}= \mathcal{f(x)(1-f(x))}$', xy=(2, 0.125), xytext=(0.6, .5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Hyperbolic tangent (tanh) function and its derivative:**

```python
def Tanh(x):
    return ((np.exp(x)-np.exp(-x))/((np.exp(x)+np.exp(-x))))

def Derivative_Tanh(x):
    return (1 - Tanh(x)**2)
```

```python
#figsize(8,5)
plt.plot(x, Tanh(x) ,'#2a9d8f', linewidth=2.5)
plt.plot(x, Derivative_Tanh(x),'r', linewidth=2)
plt.xticks(np.arange(-5,+5,1))
plt.yticks(np.arange(-1,+1,0.25))
plt.title('Hyperbolic Tangent Function', fontdict=title_font)
plt.annotate(r'$\sigma(x)=\frac{1}{1+e^{-x}}$', xy=(0, -0.25), xytext=(1, -0.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative}= 1 - f(x)^{2}$', xy=(2, 0.125), xytext=(1.5, .5), fontsize=12, arrowprops=dict(arrowstyle="->", color='b'))
```

## **ReLU function and it derivative:**

```python
def Relu(x):
    return np.maximum(0,x)

def Derivative_Relu(x):
    if x>0:
        return 1
    elif x==0:
        raise ValueError("Undefined")
    else:
        return 0
        
```

```python
y=[];
for i in x:
    y.append(Derivative_Relu(i))

#figsize(7,4)
plt.grid()
plt.plot(x, Relu(x) ,'#2a9d8f', linewidth=2.5)
plt.plot(x, y,'r', linewidth=1)
plt.xticks(np.arange(-4,+4,1))
plt.yticks(np.arange(-0.5,4,0.5))
plt.title('Relu Function', fontdict=title_font)
plt.annotate(r'$Relu(x)=\max(0,x)$', xy=(2, 2.5), xytext=(-2.7, 3.5), fontsize=12, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative}$', xy=(0, 1), xytext=(-3, 1.5), fontsize=12, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Exponential linear unit (ELU) function and its derivative:**

```python
def Elu(x, alpha):
    if x>0:
        return x
    else: 
        return alpha*(np.exp(x)-1)

def Derivative_Elu(x, alpha):
    if x<0:
        return alpha*np.exp(x)
    elif x==0 and alpha==1:
        return 1
    else:
        return 1
```

```python
y=[];
z=[];
alpha=0.5;
for i in x:
    y.append(Elu(i,alpha))
    z.append(Derivative_Elu(i,alpha))

#figsize(7,4)
plt.grid()
plt.plot(x, y ,'#2a9d8f', linewidth=2.5)
plt.plot(x, z,'r', linewidth=2)
plt.xticks(np.arange(-3,+3,1))
plt.yticks(np.arange(-0.5,4,0.5))
plt.title('Elu Function', fontdict=title_font)
plt.annotate(r'$\mathcal{Elu}$', xy=(3, 3), xytext=(0, 3.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative}$', xy=(0, 1), xytext=(-1, 2), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Softplus function and its derivative:**

```python
def Softplus(x):
    return np.log(1+np.exp(x))

def Derivative_Softplus(x):
    return 1/(1+np.exp(-x))
```

```python
#figsize(7,4)
plt.grid()
plt.plot(x, Softplus(x) ,'#2a9d8f', linewidth=2.5)
plt.plot(x, Derivative_Softplus(x),'r', linewidth=2)
plt.xticks(np.arange(-3,+3,1))
plt.yticks(np.arange(-0.5,4,0.5))
plt.title('Softplus Function', fontdict=title_font)
plt.annotate(r'$\mathcal{Softplus(x)}= \ln ({\mathcal{1}+ \mathcal{e}^{x}})$', xy=(3, 3), xytext=(0, 4.5), fontsize=12, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative} = \frac{1}{1+ \mathcal{e}^{\mathcal{-x}}}$', xy=(2, .88), xytext=(-2, 2), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Scaled exponential linear unit (SELU) function and its derivative:**

```python
def Selu(x):
    landa = 1.0507
    alpha = 1.67326
    if x<0:
        return (landa * alpha * (np.exp(x)-1))
    else:
        return landa*x

def Derivative_Selu(x):
    alpha = 1.67326
    if x<0:
        return alpha*np.exp(x)
    else:
        return 1
```

```python
y=[];
z=[];
for i in x:
    y.append(Selu(i))
    z.append(Derivative_Selu(i))

#figsize(7,4)
plt.grid()
plt.plot(x, y ,'#2a9d8f', linewidth=2.5)
plt.plot(x, z,'r', linewidth=2)
plt.xticks(np.arange(-3,+3,1))
plt.yticks(np.arange(-2,4,1))
plt.title('Selu Function', fontdict=title_font)
plt.annotate(r'$\mathcal{Selu}$', xy=(2.5, 3), xytext=(0, 3.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative}$', xy=(-1, 1), xytext=(-2, 2), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Leaky rectified linear unit (Leaky ReLU) function and its derivative:**

```python
def Leaky_Relu(x):
    return np.maximum(0.05*x,x)

def Derivative_Leaky_Relu(x):
    if x<0:
        return 0.05
    else:
        return 1
```

```python
z=[];
for i in x:
    z.append(Derivative_Leaky_Relu(i))

#figsize(7,4)
plt.grid()
plt.plot(x, Leaky_Relu(x) ,'#2a9d8f', linewidth=2.5)
plt.plot(x, z,'r', linewidth=2)
plt.xticks(np.arange(-4,4,1))
plt.yticks(np.arange(-1.5,4.5,1))
plt.title('Leaky Relu Function', fontdict=title_font)
plt.annotate(r'$\mathcal{Leaky Relu} = \mathcal{Max}(0.05\mathcal{X}, \mathcal{X})$', xy=(2, 2), xytext=(-3,3), fontsize=12, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative}$', xy=(3, 1), xytext=(3,-1), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Parametric rectified linear unit (PReLU) function and it derivative:**

```python
def PRelu(x, alpha):
    if x<0:
        return alpha*x
    else:
        return x

def Derivative_PRelu(x, alpha):
    if x<0:
        return alpha
    else:
        return 1
```

```python
y=[];
z=[];
alpha=0.2;
for i in x:
    y.append(PRelu(i, alpha))
    z.append(Derivative_PRelu(i, alpha))

#figsize(7,4)
plt.grid()
plt.plot(x, y ,'#2a9d8f', linewidth=2.5)
plt.plot(x, z,'r', linewidth=2)
plt.xticks(np.arange(-3,+3,1))
plt.yticks(np.arange(-2,4,1))
plt.title('PRelu Function', fontdict=title_font)
plt.annotate(r'$\mathcal{PRelu}$', xy=(2, 2.2), xytext=(0, 3.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative}$', xy=(0, 1), xytext=(-2, 2), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Sigmoid linear unit (SiLU) function and its derivative:**

```python
def Silu(x):
    return x/(1+np.exp(-x))

def Derivative_Silu(x):
    return ((1+np.exp(-x)+x*np.exp(-x))/((1+np.exp(-x))**2))
```

```python
#figsize(7,4)
plt.grid()
plt.plot(x, Silu(x) ,'#2a9d8f', linewidth=2.5)
plt.plot(x, Derivative_Silu(x),'r', linewidth=2)
plt.xticks(np.arange(-3,+3,1))
plt.yticks(np.arange(-0.5,4,0.5))
plt.title('Softplus Function', fontdict=title_font)
plt.annotate(r'$\mathcal{Silu(x)}=\frac{x}{1+e^{-x}}$', xy=(3, 3), xytext=(0, 4.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative} = \frac{1+ e^{\mathcal{-x}} + \mathcal{x} e^{\mathcal{-x}}}{(1+e^{\mathcal{-x}})^{2}}$', xy=(-1, .1), xytext=(-3, 1.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```

## **Gaussian function and its derivative:**

```python
def Gaussian(x):
    return np.exp(-(x**2))

def Derivative_Gaussian(x):
    return -2*x*np.exp(-(x**2))
```

```python
#figsize(7,4)
plt.grid()
plt.plot(x, Gaussian(x) ,'#2a9d8f', linewidth=2.5)
plt.plot(x, Derivative_Gaussian(x),'r', linewidth=2)
plt.xticks(np.arange(-3,+3,1))
plt.yticks(np.arange(-0.5,4,0.5))
plt.title('Gaussian Function', fontdict=title_font)
plt.annotate(r'$\mathcal{Gaussian(x)}= e^{\mathcal{-x}^{2}}$', xy=(1, 0.5), xytext=(0, 2), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
plt.annotate(r'$\mathcal{Derivative} = -2 \mathcal{x} e^{\mathcal{-x}^{2}}$', xy=(-1.6, .5), xytext=(-3.5, 1.5), fontsize=14, arrowprops=dict(arrowstyle="->", color='b'))
```
